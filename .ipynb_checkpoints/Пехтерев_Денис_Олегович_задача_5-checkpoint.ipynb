```python
import math
import numpy as np

%matplotlib notebook

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

```

```python
def f(x):
    if x == 0:
        y = 0
    else:
        y = x**2*math.log(x)
    
    return y
```

```python
x0 = 1
if x0 == 0: 
    exact = 0 # руками посчитал сам
else:
    exact = 2*x0*math.log(x0)+x0
```

```python
def f1(x, h):
    x1 = x + h
    x2 = x + 2 * h
    df = -3 / 2 * f(x) + 2 * f(x1) - 1 / 2 * f(x2)
    return df / h
```

```python
hh = np.logspace(-12, -4, 91)

plt.loglog(hh, [abs(f1(x0, h) - exact) for h in hh], 'o-') # здесь мы выставляем расхождение, \
# которое будет между вычисленным зн-м и настоящим 
```
